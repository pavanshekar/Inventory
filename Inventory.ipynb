```python
class Product:
    
    def __init__(self,price,id,quantity):
        self.price = price
        self.id = id
        self.quantity = quantity
        
    def __str__(self):
        return f"Product ID : {self.id}\tPrice : {self.price}\tQuantity : {self.quantity}"
        
    def increase_quantity(self,quantity):
        self.quantity += quantity
        
    def decrease_quantity(self,quantity):
        if self.quantity>=quantity:
            self.quantity -= quantity
            
    def total_price(self):
        return self.price*self.quantity
```

```python
class Inventory:
    
    def __init__(self,products={}):
        self.products = products
        
    def add_product(self,product):
        self.products[product.id]=product
        
    def delete_product(self,id):
        if id not in self.products.keys():
            print("Invalid id")
            return
        else:
            del self.products[id]
            
    def print_products(self):
        print("The products in the inventory are:")
        for key,value in self.products.items():
            print(value,"\tTotal Price : ",self.products[key].total_price())
    
    def number_of_items(self):
        print("The number of products : ",len(self.products))
        
    def increase_product_quantity(self,id,quantity):
        self.products[id].increase_quantity(quantity)
        
    def decrease_product_quantity(self,id,quantity):
        self.products[id].decrease_quantity(quantity)
```

```python
example = Inventory()
```

```python
while(1):
    print("\n")
    print("Inventory Management")
    print("1.Add The Product")
    print("2.Delete The Product")
    print("3.Print the Inventory Details")
    print("4.Increase Product Quantity")
    print("5.Decrease Product Quantity")
    print("6.Exit")
    print("\n")
    choice = int(input("Enter the choice : "))
    print("\n")
    if choice == 1:
        id = int(input("Enter the id : "))
        price = int(input("Enter the price : "))
        quantity = int(input("Enter quantity : "))
        p1 = Product(price,id,quantity)
        example.add_product(p1)
    elif choice == 2:
        id = int(input("Enter the id to be deleted : "))
        example.delete_product(id)
    elif choice == 3:
        example.print_products()
    elif choice == 4:
        id = int(input("Enter id : "))
        quantity = int(input("Enter quantity : "))
        example.increase_product_quantity(id,quantity)
    elif choice == 5:
        id = int(input("Enter id : "))
        quantity = int(input("Enter quantity : "))
        example.decrease_product_quantity(id,quantity)
    else:
        break
```
